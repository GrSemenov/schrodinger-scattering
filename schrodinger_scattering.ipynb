# Рассеяние волнового пакета на потенциале Юкавы

## Постановка задачи

Решаем уравнение Шрёдингера:

$$i\hbar \frac{\partial \psi}{\partial t} = \left(-\frac{\hbar^2}{2m}\nabla^2 + V(r)\right)\psi$$

**Волновой пакет** - гауссово распределение с начальным импульсом $k_0$:

$$\psi(x, 0) = A \exp\left(-\frac{(x-x_0)^2}{4\sigma^2}\right) e^{ik_0 x}$$

где:
- $x_0$ - начальное положение пакета
- $\sigma$ - пространственная ширина
- $k_0$ - начальный импульс

**Рассеивающий потенциал** - потенциал Юкавы:

$$V(r) = \frac{e^{-r/R}}{r}$$

где $R$ - радиус экранирования. При $R \to \infty$ переходит в кулоновский потенциал.

## Метод решения

Численное решение — метод **Кранка-Николсон** (неявная схема, сохраняет норму).  
Аналитическое сравнение — **борновское приближение**:

$$\frac{d\sigma}{d\Omega} = |f(\theta)|^2, \quad f(\theta) \propto \frac{1}{q^2 + R^{-2}}$$

где $q = 2k_0\sin(\theta/2)$ — переданный импульс.
